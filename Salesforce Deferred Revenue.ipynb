```python
# Define the function to determine the recognition method
def determine_recognition_method(row):
    # Safely convert columns to strings to handle potential NaNs
    payment_notes = str(row.get('Payment Notes', ''))
    product_short_name = str(row.get('Opportunity Product Short Name(s)', ''))
    salesforce_id = str(row.get('Opportunity ID', '')).strip()
    membership_start_date = row.get('Membership Start Date', None)

    # Scenario 1: Check payment notes for specific keywords
    if any(keyword.lower() in payment_notes.lower() for keyword in ['Siemens', 'Marqi', 'Flexjet', 'Buck Institute']): #Case insensitive
        return 'HLI Corporate - Payment Date Recognition'
    
    # Scenario 2: Check product short name for specific keywords
    if any(keyword in product_short_name for keyword in [
        'HLI Executive Health Program',
        'HLI Whole Body MRI Program',
        'WGS and Bloodwork'
    ]):
        return 'HLI Diagnostic Testing - Payment Date Recognition'
    
    # Scenario 3: Specific Salesforce Opportunity IDs provided by Megan Smith on 12/13/2024 (Excel doc: Delayed Payments Report 12/11/2024)
    if salesforce_id in [
        'OPP-10782', 'OPP-10783', 'OPP-11306', 'OPP-10584', 'OPP-13626', 'OPP-13627', 'OPP-14324', 'OPP-15002'
    ]:
        return 'Customer Payment Plan - Payment Date Recognition'

#    # Scenario 4: Membership Start Date is in 2023
#    if isinstance(membership_start_date, pd.Timestamp):  # Ensure valid date type
#        if membership_start_date.year == 2023:
#            return '2023 Membership Start Date & Term 17.5% / 82.5% Recognition'

#    # Scenario 5: Membership Start Date before 2023
    if isinstance(membership_start_date, pd.Timestamp):  # Ensure valid date type
        if membership_start_date.year in (2020, 2021, 2022, 2023): 
            return 'Pre-2023 12-Month Straight-Line Revenue Recognition'
        
    # Scenario 6: Default case
    return 'Membership Start Date & Term Recognition'

# Ensure necessary columns are properly typed in the DataFrame
merged_df['Payment Notes'] = merged_df['Payment Notes'].fillna('').astype(str)
merged_df['Opportunity Product Short Name(s)'] = merged_df['Opportunity Product Short Name(s)'].fillna('').astype(str)
merged_df['Salesforce Opportunity ID'] = merged_df['Salesforce Opportunity ID'].fillna('').astype(str)

# Convert Membership Start Date column to datetime (if not already)
merged_df['Membership Start Date'] = pd.to_datetime(merged_df['Membership Start Date'], errors='coerce')

# Apply the function to the DataFrame
merged_df['Revenue Recognition Method'] = merged_df.apply(determine_recognition_method, axis=1)
```

```python
# Ensure both columns are in datetime format
merged_df['Membership Start Date'] = pd.to_datetime(merged_df['Membership Start Date'], errors='coerce')
merged_df['Membership End Date'] = pd.to_datetime(merged_df['Membership End Date'], errors='coerce')

# Check for rows with invalid dates
invalid_dates = merged_df[merged_df['Membership Start Date'].isna() | merged_df['Membership End Date'].isna()]
if not invalid_dates.empty:
    print("Rows with invalid or missing dates:")
    print(invalid_dates)

# Calculate year difference for reference, skipping rows with invalid dates
merged_df['Start & End Date - Year Difference'] = (
    (merged_df['Membership End Date'] - merged_df['Membership Start Date']) / pd.Timedelta(days=365)
).round(2)

# Determine the range of months to create columns (including one month before the end date)
min_date = merged_df['Membership Start Date'].min()
max_date = merged_df['Membership End Date'].max() - pd.offsets.MonthBegin(1)

# Generate monthly date range
monthly_dates = pd.date_range(start=min_date, end=max_date, freq='MS')

# Add columns for each month in the range (if not already added)
for date in monthly_dates:
    column_name = f"{date.strftime('%B %Y')}"
    if column_name not in merged_df.columns:
        merged_df[column_name] = 0.0  # Initialize with zeros

# Distribute payments according to the specified rules
for index, row in merged_df.iterrows():
    revenue_recognition_method = row['Revenue Recognition Method']
    total_payments = row['Total Payments Received']
    term_length = row['Opportunity Term Length (Months)']
    start_date = row['Membership Start Date']
    payment_status = row['Payment Status']
    payment_1_amount = row.get('Payment 1 Amount', 0)
    payment_1_date = pd.to_datetime(row.get('Payment 1 Completed Date', None), errors='coerce')
    payment_2_amount = row.get('Payment 2 Amount', 0)
    payment_2_date = pd.to_datetime(row.get('Payment 2 Completed Date', None), errors='coerce')

    # Ensure term_length is an integer
    try:
        term_length = int(term_length)
    except ValueError:
        continue  # Skip this row if term_length is not a valid integer

    if payment_status == 'Refunded':
        exit # Do not distribute refunded payments
    elif revenue_recognition_method in ['HLI Corporate - Payment Date Recognition', 'HLI Diagnostic Testing - Payment Date Recognition', 'Customer Payment Plan - Payment Date Recognition']:
        # Recognize Payment 1 Amount
        if pd.notna(payment_1_date):
            month_column = f"{payment_1_date.strftime('%B %Y')}"
            if month_column in merged_df.columns:
                merged_df.loc[index, month_column] += payment_1_amount  # Add to existing value
                print(f"Recognized Payment 1 of {payment_1_amount} in {month_column} for row {index}")
        
        # Recognize Payment 2 Amount
        if pd.notna(payment_2_date):
            month_column = f"{payment_2_date.strftime('%B %Y')}"
            if month_column in merged_df.columns:
                merged_df.loc[index, month_column] += payment_2_amount  # Add to existing value
                print(f"Recognized Payment 2 of {payment_2_amount} in {month_column} for row {index}")

    elif revenue_recognition_method == '2023 Membership Start Date & Term 17.5% / 82.5% Recognition' and pd.notna(start_date):
        if start_date.year == 2023:  # Use Scenario 1 logic
            # 17.5% distributed over 16 months
            first_allocation = 0.175 * total_payments
            first_monthly_payment = first_allocation / 16

            # 82.5% distributed using existing logic
            remaining_allocation = 0.825 * total_payments
            if term_length == 12:
                second_monthly_payments = [remaining_allocation / 12] * 12
            elif term_length == 24:
                second_monthly_payments = [remaining_allocation * 0.6 / 12] * 12 + [remaining_allocation * 0.4 / 12] * 12
            elif term_length == 36:
                second_monthly_payments = (
                    [remaining_allocation * 0.45 / 12] * 12 +
                    [remaining_allocation * 0.28 / 12] * 12 +
                    [remaining_allocation * 0.27 / 12] * 12
                )
            elif term_length == 48:
                second_monthly_payments = (
                    [remaining_allocation * 0.33 / 12] * 12 +
                    [remaining_allocation * 0.23 / 12] * 12 +
                    [remaining_allocation * 0.22 / 12] * 12 +
                    [remaining_allocation * 0.22 / 12] * 12
                )
            elif term_length == 60:
                second_monthly_payments = (
                    [remaining_allocation * 0.25 / 12] * 12 +
                    [remaining_allocation * 0.19 / 12] * 12 +
                    [remaining_allocation * 0.19 / 12] * 12 +
                    [remaining_allocation * 0.19 / 12] * 12 +
                    [remaining_allocation * 0.18 / 12] * 12
                )
            elif term_length == 120:
                second_monthly_payments = (
                    [remaining_allocation * 0.1423 / 12] * 12 +
                    [remaining_allocation * 0.0953 / 12] * 108
                )
            else:
                second_monthly_payments = [remaining_allocation / 12] * 12 #All other payments terms distributed over 12 months (e.g. 72 months)

            # Distribute the 17.5% portion over 16 months
            current_date = start_date
            for i in range(16):  # Loop exactly 16 times
                if pd.notna(current_date):
                    month_column = f"{current_date.strftime('%B %Y')}"
                    if month_column in merged_df.columns:
                        merged_df.loc[index, month_column] += first_monthly_payment
                    current_date += pd.DateOffset(months=1)  # Increment by one month

            # Distribute the remaining 82.5% portion
            current_date = start_date
            for payment in second_monthly_payments:
                if pd.notna(current_date):
                    month_column = f"{current_date.strftime('%B %Y')}"
                    if month_column in merged_df.columns:
                        merged_df.loc[index, month_column] += payment
                    current_date += pd.DateOffset(months=1)

    elif revenue_recognition_method == 'Pre-2023 12-Month Straight-Line Revenue Recognition' and pd.notna(start_date): #Added January based on recommendation by auditors, specifically Matthew Russell
        monthly_payments = ([total_payments / 12] * 12)
        
        # Distribute payment across months
        current_date = start_date
        for payment in monthly_payments:
            if pd.notna(current_date):
                month_column = f"{current_date.strftime('%B %Y')}"
                if month_column in merged_df.columns:
                    merged_df.loc[index, month_column] += payment
                current_date += pd.DateOffset(months=1)
        
    elif revenue_recognition_method == 'Membership Start Date & Term Recognition' and pd.notna(start_date):
        # Use Scenario 2 logic
        if term_length == 12:
            monthly_payments = [total_payments / 12] * 12
        elif term_length == 24:
            monthly_payments = [total_payments * 0.6 / 12] * 12 + [total_payments * 0.4 / 12] * 12
        elif term_length == 36:
            monthly_payments = (
                [total_payments * 0.45 / 12] * 12 +
                [total_payments * 0.28 / 12] * 12 +
                [total_payments * 0.27 / 12] * 12
            )
        elif term_length == 48:
            monthly_payments = (
                [total_payments * 0.33 / 12] * 12 +
                [total_payments * 0.23 / 12] * 12 +
                [total_payments * 0.22 / 12] * 12 +
                [total_payments * 0.22 / 12] * 12
            )
        elif term_length == 60:
            monthly_payments = (
                [total_payments * 0.25 / 12] * 12 +
                [total_payments * 0.19 / 12] * 12 +
                [total_payments * 0.19 / 12] * 12 +
                [total_payments * 0.19 / 12] * 12 +
                [total_payments * 0.18 / 12] * 12
            )
        elif term_length == 120: #Rounding issue with current numbers results in 99.96% of payments are distributed
            monthly_payments = (
                [total_payments * 0.1423 / 12] * 12 + #Updated from .1419 to .1423 (December 2024)
                [total_payments * 0.0953 / 12] * 108
            )
        else:
            monthly_payments = [total_payments / 12] * 12 #All other payments terms distributed over 12 months (e.g. 72 months)

        # Distribute payment across months
        current_date = start_date
        for payment in monthly_payments:
            if pd.notna(current_date):
                month_column = f"{current_date.strftime('%B %Y')}"
                if month_column in merged_df.columns:
                    merged_df.loc[index, month_column] += payment
                current_date += pd.DateOffset(months=1)
```

```python
# Convert date columns to datetime
sales_commission_df['Payment 1 Completed Date'] = pd.to_datetime(sales_commission_df['Payment 1 Completed Date'])
sales_commission_df['Payment 2 Completed Date'] = pd.to_datetime(sales_commission_df['Payment 2 Completed Date'])

# Function to get the quarter and year from a datetime object
def get_quarter_year(date):
    if pd.isna(date):
        return ''
    return f"Q{date.quarter} {date.year}"

# Apply the function to get quarters and years for each date column
sales_commission_df['Payment 1 Quarter Year'] = sales_commission_df['Payment 1 Completed Date'].apply(get_quarter_year)
sales_commission_df['Payment 2 Quarter Year'] = sales_commission_df['Payment 2 Completed Date'].apply(get_quarter_year)

# Concatenate the quarters and years into the new column 'Sales Commission Quarter(s)'
def concatenate_quarters_years(row):
    quarters_years = {row['Payment 1 Quarter Year'], row['Payment 2 Quarter Year']}
    return ', '.join(sorted(filter(None, quarters_years)))  # Remove empty strings and sort for consistency

sales_commission_df['Sales Commission Quarter(s)'] = sales_commission_df.apply(concatenate_quarters_years, axis=1)

# Define the year and quarter to sum payments for
specified_year = 2024
specified_quarter = 4

# Create dynamic column title
dynamic_column_title = f'Q{specified_quarter} {specified_year} Payment Totals'

# Function to sum payments for a given year and quarter
def sum_payments_by_period(df, year, quarter):
    # Create a period for the specified year and quarter
    period = f"Q{quarter} {year}"
    
    # Filter rows for the specified year and quarter
    mask1 = df['Payment 1 Quarter Year'] == period
    mask2 = df['Payment 2 Quarter Year'] == period
    
    # Sum payments based on the specified period
    payment_1_sum = df.loc[mask1, 'Payment 1 Amount'].sum()
    payment_2_sum = df.loc[mask2, 'Payment 2 Amount'].sum()
    
    # Return the total sum of payments
    return payment_1_sum + payment_2_sum

# Apply function to get total payments for each row based on the specified year and quarter
def calculate_quarter_year_payment_totals(row):
    # Compute the total payment based on the precomputed periods and specified variables
    total_payments = 0
    for period in [row['Payment 1 Quarter Year']]:
        if period == f"Q{specified_quarter} {specified_year}":
            total_payments += row['Payment 1 Amount'] if row['Payment 1 Quarter Year'] == period else 0

    for period in [row['Payment 2 Quarter Year']]:
        if period == f"Q{specified_quarter} {specified_year}":
            total_payments += row['Payment 2 Amount'] if row['Payment 2 Quarter Year'] == period else 0
    return total_payments

# Create the new column with the dynamic title based on specified variables
sales_commission_df[dynamic_column_title] = sales_commission_df.apply(calculate_quarter_year_payment_totals, axis=1)

# Drop intermediate columns if not needed
sales_commission_df.drop(columns=['Payment 1 Quarter Year', 'Payment 2 Quarter Year'], inplace=True)

# Convert back to date (without time)
sales_commission_df['Payment 1 Completed Date'] = sales_commission_df['Payment 1 Completed Date'].dt.date
sales_commission_df['Payment 2 Completed Date'] = sales_commission_df['Payment 2 Completed Date'].dt.date
```
