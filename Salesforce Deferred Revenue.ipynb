### Add payments and defer across term length

```python
# Convert date columns to datetime format
merged_df['Membership Start Date'] = pd.to_datetime(merged_df['Membership Start Date'])
merged_df['Membership End Date'] = pd.to_datetime(merged_df['Membership End Date'])

merged_df['Start & End Date - Year Difference'] = (merged_df['Membership End Date'] - merged_df['Membership Start Date']) / pd.Timedelta(days=365)
merged_df['Start & End Date - Year Difference'] = merged_df['Start & End Date - Year Difference'].round(2)



# Determine the range of months to create columns (including one month before the end date)
min_date = merged_df['Membership Start Date'].min()
max_date = merged_df['Membership End Date'].max() - pd.offsets.MonthBegin(1)

# Generate monthly date range
monthly_dates = pd.date_range(start=min_date, end=max_date, freq='MS')

# Add columns for each month in the range (if not already added)
for date in monthly_dates:
    column_name = f"{date.strftime('%B %Y')}"
    if column_name not in merged_df.columns:
        merged_df[column_name] = 0.0  # Initialize with zeros

# Reference date for condition
condition_date = datetime(2024, 1, 1) #When new reference schedule was effective, per Ellen on 6/18/2024

# Distribute payments according to the specified rules
for index, row in merged_df.iterrows():
    total_payments = row['Total Payments Received']
    term_length = row['Opportunity Term Length (Months)']
    start_date = row['Membership Start Date']
    end_date = row['Membership End Date']
    
    # Ensure term_length is an integer
    try:
        term_length = int(term_length)
    except ValueError:
        continue  # Skip this row if term_length is not a valid integer
    
    # Initialize monthly_payments with a default empty list
    monthly_payments = []

    # Calculate monthly payments based on term length if start date is on or after 2024-01-01
    if start_date >= condition_date:
        if term_length == 12:
            monthly_payments = [total_payments / 12] * 12
        elif term_length == 24:
            monthly_payments = [total_payments * 0.6 / 12] * 12 + [total_payments * 0.4 / 12] * 12
        elif term_length == 36:
            monthly_payments = [total_payments * 0.45 / 12] * 12 + [total_payments * 0.28 / 12] * 12 + [total_payments * 0.27 / 12] * 12
        elif term_length == 48:
            monthly_payments = [total_payments * 0.33 / 12] * 12 + [total_payments * 0.23 / 12] * 12 + [total_payments * 0.22 / 12] * 12 + [total_payments * 0.22 / 12] * 12
        elif term_length == 60:
            monthly_payments = [total_payments * 0.25 / 12] * 12 + [total_payments * 0.19 / 12] * 12 + [total_payments * 0.19 / 12] * 12 + [total_payments * 0.19 / 12] * 12 + [total_payments * 0.18 / 12] * 12 
        elif term_length == 120:
            monthly_payments = [total_payments * 0.1419 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12
    else:
        # Distribute payments evenly across 12 months if the start date is BEFORE 2024-01-01
        monthly_payments = [total_payments / 12] * 12

    # Distribute payment across months
    current_date = start_date
    for payment in monthly_payments:
        if current_date <= end_date:
            month_column = f"{current_date.strftime('%B %Y')}"
            if month_column in merged_df.columns:
                merged_df.loc[index, month_column] += payment
            current_date = current_date + pd.offsets.MonthEnd(0) + timedelta(days=1)
        else:
            break
```
