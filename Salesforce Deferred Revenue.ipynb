### Add payments and defer across term length

```python
# Convert date columns to datetime format
merged_df['Membership Start Date'] = pd.to_datetime(merged_df['Membership Start Date'])
merged_df['Membership End Date'] = pd.to_datetime(merged_df['Membership End Date'])

merged_df['Start & End Date - Year Difference'] = (merged_df['Membership End Date'] - merged_df['Membership Start Date']) / pd.Timedelta(days=365)
merged_df['Start & End Date - Year Difference'] = merged_df['Start & End Date - Year Difference'].round(2)



# Determine the range of months to create columns (including one month before the end date)
min_date = merged_df['Membership Start Date'].min()
max_date = merged_df['Membership End Date'].max() - pd.offsets.MonthBegin(1)

# Generate monthly date range
monthly_dates = pd.date_range(start=min_date, end=max_date, freq='MS')

# Add columns for each month in the range (if not already added)
for date in monthly_dates:
    column_name = f"{date.strftime('%B %Y')}"
    if column_name not in merged_df.columns:
        merged_df[column_name] = 0.0  # Initialize with zeros

# Reference date for condition
condition_date = datetime(2024, 1, 1) #When new reference schedule was effective, per Ellen on 6/18/2024

# Distribute payments according to the specified rules
for index, row in merged_df.iterrows():
    total_payments = row['Total Payments Received']
    term_length = row['Opportunity Term Length (Months)']
    start_date = row['Membership Start Date']
    end_date = row['Membership End Date']
    
    # Ensure term_length is an integer
    try:
        term_length = int(term_length)
    except ValueError:
        continue  # Skip this row if term_length is not a valid integer
    
    # Initialize monthly_payments with a default empty list
    monthly_payments = []

    # Calculate monthly payments based on term length if start date is on or after 2024-01-01
    if start_date >= condition_date:
        if term_length == 12:
            monthly_payments = [total_payments / 12] * 12
        elif term_length == 24:
            monthly_payments = [total_payments * 0.6 / 12] * 12 + [total_payments * 0.4 / 12] * 12
        elif term_length == 36:
            monthly_payments = [total_payments * 0.45 / 12] * 12 + [total_payments * 0.28 / 12] * 12 + [total_payments * 0.27 / 12] * 12
        elif term_length == 48:
            monthly_payments = [total_payments * 0.33 / 12] * 12 + [total_payments * 0.23 / 12] * 12 + [total_payments * 0.22 / 12] * 12 + [total_payments * 0.22 / 12] * 12
        elif term_length == 60:
            monthly_payments = [total_payments * 0.25 / 12] * 12 + [total_payments * 0.19 / 12] * 12 + [total_payments * 0.19 / 12] * 12 + [total_payments * 0.19 / 12] * 12 + [total_payments * 0.18 / 12] * 12 
        elif term_length == 120:
            monthly_payments = [total_payments * 0.1419 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12 + [total_payments * 0.0953 / 12] * 12
    else:
        # Distribute payments evenly across 12 months if the start date is BEFORE 2024-01-01
        monthly_payments = [total_payments / 12] * 12

    # Distribute payment across months
    current_date = start_date
    for payment in monthly_payments:
        if current_date <= end_date:
            month_column = f"{current_date.strftime('%B %Y')}"
            if month_column in merged_df.columns:
                merged_df.loc[index, month_column] += payment
            current_date = current_date + pd.offsets.MonthEnd(0) + timedelta(days=1)
        else:
            break
```

```python
# Convert 'ADHOC Payment Date' columns to datetime format
date_columns = [col for col in merged_df.columns if col.startswith('ADHOC Payment Date')]
for col in date_columns:
    merged_df[col] = pd.to_datetime(merged_df[col])

# Ensure the date columns are in datetime format
for i in range(1, 4):
    date_col = f'ADHOC Payment Date {i}'
    merged_df[date_col] = pd.to_datetime(merged_df[date_col], errors='coerce')

# Define the function to update monthly payments
def update_monthly_payments(row):
    for i in range(1, 4):  # Adjust range based on the number of sets (in this example, up to 3)
        amount_col = f'ADHOC Payment Amount {i}'
        date_col = f'ADHOC Payment Date {i}'
        
        if pd.notnull(row[amount_col]) and row[amount_col] > 0:
            payment_date = row[date_col]
            if pd.notnull(payment_date):
                month_year = payment_date.strftime('%B %Y')
                col_name = f'{month_year} - ADHOC'
                
                if col_name not in merged_df.columns:
                    merged_df[col_name] = 0
                
                merged_df.at[row.name, col_name] += row[amount_col]

# Apply the function to each row in the DataFrame
merged_df.apply(update_monthly_payments, axis=1)

# Optionally, convert 'ADHOC Payment Date' columns back to string format for display/export
for col in date_columns:
    merged_df[col] = merged_df[col].dt.strftime('%Y-%m-%d')
```

```python
# Convert date columns to datetime
sales_commission_df['Payment 1 Completed Date'] = pd.to_datetime(sales_commission_df['Payment 1 Completed Date'])
sales_commission_df['Payment 2 Completed Date'] = pd.to_datetime(sales_commission_df['Payment 2 Completed Date'])

# Function to get the quarter and year from a datetime object
def get_quarter_year(date):
    if pd.isna(date):
        return ''
    return f"Q{date.quarter} {date.year}"

# Apply the function to get quarters and years for each date column
sales_commission_df['Payment 1 Quarter Year'] = sales_commission_df['Payment 1 Completed Date'].apply(get_quarter_year)
sales_commission_df['Payment 2 Quarter Year'] = sales_commission_df['Payment 2 Completed Date'].apply(get_quarter_year)

# Concatenate the quarters and years into the new column 'Sales Commission Quarter(s)'
def concatenate_quarters_years(row):
    quarters_years = {row['Payment 1 Quarter Year'], row['Payment 2 Quarter Year']}
    return ', '.join(sorted(filter(None, quarters_years)))  # Remove empty strings and sort for consistency

sales_commission_df['Sales Commission Quarter(s)'] = sales_commission_df.apply(concatenate_quarters_years, axis=1)

# Define the year and quarter to sum payments for
specified_year = 2024
specified_quarter = 2  # For Q2

# Create dynamic column title
dynamic_column_title = f'Q{specified_quarter} {specified_year} Payment Totals'

# Function to sum payments for a given year and quarter
def sum_payments_by_period(df, year, quarter):
    # Create a period for the specified year and quarter
    period = f"Q{quarter} {year}"
    
    # Filter rows for the specified year and quarter
    mask1 = df['Payment 1 Quarter Year'] == period
    mask2 = df['Payment 2 Quarter Year'] == period
    
    # Sum payments based on the specified period
    payment_1_sum = df.loc[mask1, 'Payment 1 Amount'].sum()
    payment_2_sum = df.loc[mask2, 'Payment 2 Amount'].sum()
    
    # Return the total sum of payments
    return payment_1_sum + payment_2_sum

# Apply function to get total payments for each row based on the specified year and quarter
def calculate_quarter_year_payment_totals(row):
    # Compute the total payment based on the precomputed periods and specified variables
    total_payments = 0
    for period in [row['Payment 1 Quarter Year']]:
        if period == f"Q{specified_quarter} {specified_year}":
            total_payments += row['Payment 1 Amount'] if row['Payment 1 Quarter Year'] == period else 0

    for period in [row['Payment 2 Quarter Year']]:
        if period == f"Q{specified_quarter} {specified_year}":
            total_payments += row['Payment 2 Amount'] if row['Payment 2 Quarter Year'] == period else 0
    return total_payments

# Create the new column with the dynamic title based on specified variables
sales_commission_df[dynamic_column_title] = sales_commission_df.apply(calculate_quarter_year_payment_totals, axis=1)

# Drop intermediate columns if not needed
sales_commission_df.drop(columns=['Payment 1 Quarter Year', 'Payment 2 Quarter Year'], inplace=True)

# Convert back to date (without time)
sales_commission_df['Payment 1 Completed Date'] = sales_commission_df['Payment 1 Completed Date'].dt.date
sales_commission_df['Payment 2 Completed Date'] = sales_commission_df['Payment 2 Completed Date'].dt.date
```

```python
# Define initial commission amounts
full_commission = 300
medium_range_commission = 0.75 * full_commission
low_range_commission = 0.25 * full_commission
additional_yearly_commission = 150

# Function to calculate the initial earned commission
def calculate_commission(row):
    base_commission = 0
    if 'Initial' in row['Opportunity Product Short Name(s)'] and row[dynamic_column_title] > 0:
        # Calculate base commission based on payment ranges
        if 10000 <= row[dynamic_column_title] / row['Opportunity Term Length (Years)'] <= 14999:
            base_commission = medium_range_commission
        elif row[dynamic_column_title] / row['Opportunity Term Length (Years)'] <= 9999:
            base_commission = low_range_commission
        else:
            base_commission = full_commission
        
        # Add commission for each additional year beyond the first
        additional_commission = max(0, row['Opportunity Term Length (Years)'] - 1) * additional_yearly_commission
        return base_commission + additional_commission
    
    return 0  # No commission if 'Initial' is not in the product short name

# Apply the function to create the 'Initial Earned Commission' column
sales_commission_df['Initial Earned Commission'] = sales_commission_df.apply(calculate_commission, axis=1)
```

```python
# Define renewal commission amounts 
commission_per_year = 150

# Function to calculate the renewal earned commission
def calculate_renewal_commission(row):
    if 'Renewal' in row['Opportunity Product Short Name(s)'] and row[dynamic_column_title] > 0:
        return row['Opportunity Term Length (Years)'] * commission_per_year
    return 0  # No commission if 'Renewal' is not in the product short name

# Apply the function to create the 'Renewal Earned Commission' column
sales_commission_df['Renewal Earned Commission'] = sales_commission_df.apply(calculate_renewal_commission, axis=1)
```

```python
# Define commission amounts
upgrade_commission = 200

# Function to calculate the earned commission
def calculate_commission(row):
    commission = 0
    if 'Upgrade' in row['Opportunity Product Short Name(s)'] and row[dynamic_column_title] > 0:
        commission += upgrade_commission
        commission += (row['Opportunity Term Length (Years)']-1) * commission_per_year #multi-year bonus additions for 'upgrades'
    return commission

# Apply the function to create the 'Earned Commission' column
sales_commission_df['Upgrade Earned Commission'] = sales_commission_df.apply(calculate_commission, axis=1)
```

```python
sales_commission_df['Total Earned Commission'] = sales_commission_df['Initial Earned Commission'] + sales_commission_df['Renewal Earned Commission'] + sales_commission_df['Upgrade Earned Commission']
```
